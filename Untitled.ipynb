```python
def converttostr(input_seq, seperator):
   # Join all the strings in list
   final_str = seperator.join(input_seq)
   return final_str
```

```python
def process_list(input_list_object):
    iRow = []
    company_split = input_list_object.lstrip().split("\n")[0]
    
    org_list = input_list_object.lstrip().split("\n")
    if org_list[0] == company_split:
        del org_list[0]
    
    for org in org_list:
        strip_org = org.strip()
        split_org = org.split(" ")
        for o in split_org:
            o_strip = o.strip()
            if len(o_strip) > 0:
                cap = 0
                split_index = 0
                split_strings = []
                for c in o_strip:
                    if c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" and not o_strip.isupper():
                        if cap != 0:
                            split_string = o_strip[:cap] + " " + o_strip[cap:]
                            split_strings.append(split_string)
                            #split_strings.append(split_string.replace(".", " "))
                            split_index +=cap
                    cap +=1
                if len(split_strings) > 0:
                    iRow.append(split_strings[len(split_strings) - 1])
                if split_index == 0:
                    iRow.append(o_strip)
                    
    return converttostr(iRow, " "), company_split
```

```python
def process_contact(input_object):
    iRow = []
    split_object = input_object.lstrip().split("\n")
    company_split = split_object[0].strip() + " " + split_object[1].strip()
    
    for org in split_object:
        org_strip = org.strip()
        if org_strip not in company_split and ":" not in org:
            split_org = org.split(" ")
            for o in split_org:
                o_strip = o.strip()
                if len(o_strip) > 0:
                    cap = 0
                    split_index = 0
                    split_strings = []
                    for c in o_strip:
                        if c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ" and not o_strip.isupper():
                            if cap != 0:
                                split_string = o_strip[:cap] + " " + o_strip[cap:]
                                split_strings.append(split_string.replace(".", " "))
                                split_index +=cap
                        cap +=1
                    if len(split_strings) > 0:
                        iRow.append(split_strings[len(split_strings) - 1])
                    if split_index == 0:
                        iRow.append(o_strip.replace(".", " "))
    
    return converttostr(iRow, " "), company_split
```

```python
delegation_records = {}
delegation_list = []

for row in df.iterrows():
    sub_url = url + "/" + row[1]['Domain'][1:] + ".html"
    print(sub_url)
    try:
        page = requests.get(sub_url, timeout=5)
        doc = lh.fromstring(page.content)
        del_record = str(doc.get_element_by_id("main_right").text_content())

        try:
            data_list = []
            
            if row[1]['Type'] in ['generic', 'sponsored']:
                sponsoring_org = del_record.split("Sponsoring Organisation")[1].split("Administrative Contact")[0]
            else:
                sponsoring_org = del_record.split("ccTLD Manager")[1].split("Administrative Contact")[0]

            administrative_contact = del_record.split("Administrative Contact")[1].split("Technical Contact")[0]
            technical_contact = del_record.split("Technical Contact")[1].split("Name Servers")[0]
            clean_sponsoring_org, sponsoring_co = process_list(sponsoring_org)
            clean_administrative_contact, adminstrative_co = process_contact(administrative_contact)
            clean_technical_contact, technical_co = process_contact(technical_contact)
            
            try:
                last_updated_str = del_record.split("Record last updated")[1].split("Registration date")[0].replace(".", "").strip()
                registration_date_str = del_record.split("Registration date")[1].split("\n")[0].replace(".", "").strip()

                date_format = '%Y-%m-%d'
                last_updated_date = datetime.strptime(last_updated_str, date_format)
                registration_date = datetime.strptime(registration_date_str, date_format)
            except:
                print(f"Error processing dates for {sub_url}")

            sponsoring_org_dict = {"domain": row[1]['Domain'],
                                   "type":row[1]['Type'],
                                   "organization-type":"Sponsoring Organization",
                                   "organization-name":sponsoring_co,
                                   "organization-address":clean_sponsoring_org,
                                   "updated": last_updated_date,
                                   "registered": registration_date}
            
            delegation_list.append(sponsoring_org_dict)
            
            adminstrative_contact_dict = {"domain": row[1]['Domain'],
                                          "type":row[1]['Type'],
                                          "organization-type":"Administrative Contact",
                                          "organization-name":adminstrative_co,
                                          "organization-address":clean_administrative_contact,
                                          "updated": last_updated_date,
                                          "registered": registration_date}
            delegation_list.append(adminstrative_contact_dict)
            
            technical_contact_dict = {"domain": row[1]['Domain'],
                                      "type":row[1]['Type'],
                                      "organization-type":"Technical Contact",
                                      "organization-name":technical_co,
                                      "organization-address":clean_technical_contact,
                                      "updated": last_updated_date,
                                      "registered": registration_date}
            
            delegation_list.append(technical_contact_dict)

        except:
            print(f"Error processing {sub_url}")
    except requests.exceptions.Timeout:
        print(f"Timeout reached on {sub_url}")
    except (IndexError, KeyError):
        print(f"{sub_url} does not contain required elements")
```

```python
import json
def geocode_address(address=None):
    """Use World Geocoder to get XY for one address at a time."""
    querystring = {
        "f": "json",
        "singleLine": address}
    url = "https://geocode.arcgis.com/arcgis/rest/services/World/GeocodeServer/findAddressCandidates"  # noqa: E501
    response = requests.request("GET", url, params=querystring)
    p = response.text
    j = json.loads(p)
    location = j['candidates'][0]['location']  # returns first location as X, Y
    return location
```
